```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# x values
x = np.linspace(-3, 3, 15)
y = 0.5*x**5 - 2*x**3 + x + np.random.randn(15)*10
X = []
Y = []


for i in x:
    X.append([i])
    
for i in y:
    Y.append([i])
    



# y values (non-linear relationship)


```

```python
y
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
plt.scatter(x, y, label="Actual")
```

```python
i = 1
poly = PolynomialFeatures(i)
x_poly = poly.fit_transform(X)
```

```python
x_poly
```

```python
scaler = StandardScaler()
x_poly_stnd = scaler.fit_transform(x_poly)
```

```python
x_poly_stnd
```

```python
model = LinearRegression()
model.fit(x_poly_stnd, Y)
output = model.predict(x_poly_stnd)
print(output)

fig = plt.figure()
plt.scatter(X, Y)
plt.scatter(X, output)
plt.show()
```

```python
for i in range(1, 10):
    poly = PolynomialFeatures(i)
    x_poly = poly.fit_transform(X)
    scaler = StandardScaler()
    x_poly_stnd = scaler.fit_transform(x_poly)
    model = LinearRegression()
    model.fit(x_poly_stnd, Y)
    output = model.predict(x_poly_stnd)

    fig = plt.figure()
    plt.scatter(X, Y)
    plt.plot(X, output)
    plt.show()
```
